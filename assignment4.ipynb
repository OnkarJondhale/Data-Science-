```python
import pandas as pd
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import warnings 
warnings.filterwarnings('ignore')
```

```python
mean=100
std_deviation=4
datasize=100000
```

creating a dataset with random numbers from datasize.

```python
df=np.random.normal(mean,std_deviation,datasize)
```

Ploting a histogram to show normal distribution.

```python
plt.figure(figsize=(8,6))
sns.histplot(df,bins=50,kde=True,color='green',stat='density',edgecolor='black',alpha=0.6)
sns.kdeplot(df, color='red', linewidth=2)

plt.title('Normal Distribution with Bell Curve (mean=100, std_deviation=4, size=100000)',fontsize=12)
plt.xlabel('Value',fontsize=10)
plt.ylabel('Density',fontsize=10)

plt.show()
```

The smooth curve ensures normal distribution across the dataset. As the datasize is large so we get to see a perfect bell shape with smooth curve.
