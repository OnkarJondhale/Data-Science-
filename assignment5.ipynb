```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('data.csv', encoding='windows-1252')
df
```

```python
df.head()
```

```python
df.tail()
```

```python
df.isna()
```

```python
df.isna().sum()
```

```python
md = df['pm2_5'].median()
md
```

```python
df['pm2_5'] = df['pm2_5'].fillna(md);
```

```python
df
```

```python
df.isna().sum()
```

```python
spm_md = df['spm'].median()
spm_md
df['spm'] = df['spm'].fillna(spm_md);
df['so2'] = df['so2'].fillna(df['so2'].median());
df['no2'] = df['no2'].fillna(df['no2'].median());
df['rspm'] = df['rspm'].fillna(df['rspm'].median());
```

```python
df.isna().sum()
```

```python
df.drop('agency',axis=1,inplace=True)
```

```python
df.drop('location_monitoring_station',axis=1,inplace=True)
df
```

```python
df.dropna(subset=["stn_code"],inplace=True)
```

```python
df.isna().sum()
```

```python
df.dropna(subset=["type"],inplace=True)
```

```python
df.dropna(subset=["date"],inplace=True)
```

```python
df
```

```python
df.info()
```
