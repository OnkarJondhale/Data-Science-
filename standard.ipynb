```python
# check for null or missing values
df.isna().sum()
```

```python
# check for null or missing values if they are removed
df.dropna();
```

```python
# again check for null values after removal it should be 0
df.isna().sum()
```

```python
# Create a new column filled with mean if missing values are present
df['bmi_mean_filled'] = df['bmi'].fillna(df['bmi'].mean())
df
# Create a new column filled with median if missing values are present
df['bmi_median_filled'] = df['bmi'].fillna(df['bmi'].median())
df
# Create a new column filled with mode if missing values are present
df['bmi_mode_filled'] = df['bmi'].fillna(df['bmi'].mode()[0])
df
```

```python
# check for duplicate values
df.duplicated()
```

```python
# get duplicate values count
df.duplicated().sum()
```

```python
# drop duplicate values
df.drop_duplicates()
```

```python
# get duplicate values count
df.duplicated().sum()
```

```python
# handle outliers
q1 = df['bmi'].quantile(0.25)
q3 = df['bmi'].quantile(0.75)
iqr = (q3-q1) * 1.5
lb = q1-iqr
ub = q3+iqr
print("q1 :",q1,"q3 :",q3,"iqr :",iqr,"lb :",lb,"ub :",ub)
```

```python
# visualize the outliers
plt.boxplot(df['bmi']) 
plt.title('Boxplot of BMI') 
plt.ylabel('BMI') 
plt.show()
```

```python
# remove the outliers from the dataset
df_removed_outliers = df[(df['charges']< lb) | (df['charges']> ub)]
```

```python
# data transformation
```

```python
# minimax normalization
df_normalized = (df['bmi']-df['bmi'].min())/(df['bmi'].max()-df['bmi'].min())
df_normalized
```

```python
# using z-score
df_normalized = (df['charges']-df['charges'].mean())/(df['charges'].std())
df_normalized
```

```python
# binning
bins = [0,18,35,50,70]
labels = ["child","young","mid-age","old"]
df["age_group"] = pd.cut(df['age'],bins = bins,labels = labels,right=False)
df
```

```python
df_encoded = pd.get_dummies(df, columns=['sex'], drop_first=True)
df_encoded['sex_male'] = df_encoded['sex_male'].astype(int)
df_encoded
```

```python
df_encoded = pd.get_dummies(df, columns=['region'], drop_first=True)
df_encoded['region_northwest'] = df_encoded['region_northwest'].astype(int)
df_encoded['region_southeast'] = df_encoded['region_southeast'].astype(int)
df_encoded['region_southwest'] = df_encoded['region_northwest'].astype(int)
df_encoded
```

```python
# label encoding
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
df['smoker'] = label_encoder.fit_transform(df['smoker'])
df
```

```python
# Covarience
numeric_df = df.select_dtypes(include=['number'])
cov = numeric_df.cov()
cov
```
